```python
import numpy as np
```

```python
def updateX(matriz, lat, x, x_i, y_i):
    try:
        matriz[x_i+1, y_i] = lat
    except:
        pass
    return x
    
def updateY(matriz, lat, y, x_i, y_i):
    try:
        matriz[x_i, y_i+1] = lat
    except:
        pass
    return y
def updateLatX(matriz, lat, x_i, y_i):
    try:
        matriz[x_i+3, y_i] = lat
    except:
        pass
def updateLatY(matriz, lat, x_i, y_i):
    try:
        matriz[x_i, y_i+3] = lat
    except:
        pass
```

```python
matriz = np.zeros((49, 49))
matriz.shape
x, y, ctr, lat = 3, 3, -256, 64
for x_i, line in enumerate(matriz):
    x -= 1
    for y_i, value in enumerate(line):    
        # define valor central
        if x_i == y_i:
            y -= 1
            matriz[x_i, y_i] = -256
            x = updateX(matriz, lat, x, x_i, y_i) if x > 0 else 3
            y = updateY(matriz, lat, y, x_i, y_i) if y > 0  else 3
            updateLatX(matriz, lat, x_i, y_i)
            updateLatY(matriz, lat, x_i, y_i)
matriz
```

```python
open('matriz.txt', 'w').close()
f = open("matriz.txt", "a")
for line in matriz:
    st = ''
    for value in line:
        st = st + str(int(value)) + ' '
    print(st)
    f.write(st+'\n')
f.close()
```
